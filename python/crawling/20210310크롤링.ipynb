```python
products = []
dates = []
texts = []
stars = []
cnt=1   #리뷰index
page=1
while True:
    j=1
    print ("페이지", page ,"\n") 
    time.sleep(2)
    while True: #한페이지에 20개의 리뷰, 마지막 리뷰에서 error발생
        try:
            date = driver.find_element_by_xpath('//*[@id="REVIEW"]/div/div[3]/div/div[2]/ul/li['+str(j)+']/div/div/div/div[1]/div/div[1]/div[1]/div[2]/div[2]/span[1]').text
            dates.append(date)
            text = driver.find_element_by_xpath('(//*[@id="REVIEW"]/div/div[3]/div/div[2]/ul/li['+str(j)+']/div/div[1]/div/div[1]/div/div/div[2]/div/span)[last()]').text
            texts.append(text)
            star = driver.find_element_by_xpath('//*[@id="REVIEW"]/div/div[3]/div/div[2]/ul/li['+str(j)+']/div/div/div/div[1]/div/div[1]/div[1]/div[2]/div[1]/em').text
            stars.append(star)
            try:
                product = driver.find_element_by_xpath('//*[@id="REVIEW"]/div/div[3]/div/div[2]/ul/li['+str(j)+']/div/div[1]/div/div[1]/div/div/div[1]/div[2]/div[3]/div/button/span[1]').text
            except :
                product = 'null'
            products.append(product)
            
            j+=1
            print(cnt, product ,date,text, "\n")
            cnt+=1 
        except: break
            
    time.sleep(2)
    
    
    if page<11:#page10까지 적용
        try: #리뷰의 마지막 페이지에서 error발생
            next_page=driver.find_element_by_xpath('//*[@id="REVIEW"]/div/div[3]/div/div[2]/div/div/a['+str(page+2)+']').click() 
            page +=1
            print("test2")
        except: break #리뷰의 마지막 페이지에서 process 종료 
          
    else :
        try: #page11부터
            if page%10==0: next_page=driver.find_element_by_xpath('//*[@id="REVIEW"]/div/div[3]/div/div[2]/div/div/a[12]').click()
            else : next_page=driver.find_element_by_xpath('//*[@id="REVIEW"]/div/div[3]/div/div[2]/div/div/a['+str(page%10+2)+']').click()
            page+=1
            print("test3")
        except: break 
            


```
